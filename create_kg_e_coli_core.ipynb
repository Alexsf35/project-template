```python
from pathlib import Path

# 1. Importar o BioCypher
from biocypher import BioCypher

# 2. Importar o teu Adapter (ajusta o caminho do import consoante a tua estrutura de pastas)
# Assumindo que o notebook está na raiz do project-template e o adapter na pasta template_package/adapters
from template_package.adapters.gsmm_adapter import GSMMAdapter 

print("A inicializar...")

# Parâmetros
sbml_fp = Path("models/e_coli_core.xml")

# 3. Inicializar o BioCypher
# Ele vai ler as configurações dos ficheiros yaml na pasta config/
bc = BioCypher(
    biocypher_config_path="config/biocypher_config.yaml",
    schema_config_path="config/schema_config.yaml"
)

# 4. Instanciar o Adaptador
print(f"A ler o modelo {sbml_fp}...")
adapter = GSMMAdapter(sbml_paths=str(sbml_fp), cache_in_memory=True)

# 5. Entregar os dados ao BioCypher para ele escrever no formato correto!
print("A gerar e a gravar os Nós...")
bc.write_nodes(adapter.get_nodes())

print("A gerar e a gravar as Arestas...")
bc.write_edges(adapter.get_edges())

# 6. Gerar o script de importação para o Neo4j
print("A criar os ficheiros finais...")
bc.write_import_call()

#bc.summary()
print("Concluído! Dados guardados na pasta biocypher-out/")
```
